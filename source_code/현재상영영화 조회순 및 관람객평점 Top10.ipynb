### 오늘날짜 기준 - 날짜별 영화별 평점 찾기

```python
date = pd.date_range('2020-1-1', periods = 79, freq = 'D')
date
```

#### 2020년 1월부터 반복문을 시행하기에 앞서 오늘자(어제자)의 영화만을 가지고 이름과 관람객평점을 가져와본다.

```python
movie_name = []

for n in range(0,30):
    movie_name.append(soup.find_all('div', 'tit3')[n].a.string)
    
movie_name
```

```python
movie_point = []

for n in tqdm_notebook(range(0,30)) : 
    html = urlopen(URL[n])
    soup = BeautifulSoup(html, 'html.parser')
    soup = soup.find_all('div', 'star_score')
    
    if soup[0].find(class_ = 'st_off').span.string[-5:-1] == '평점 없' :
        movie_point.append(np.nan)
    elif soup[0].find(class_ = 'st_off').span.string[-5:-1] == '10점 ':
        movie_point.append('10')
    else :
        movie_point.append(soup[0].find(class_ = 'st_off').span.string[-5:-1])

movie_point
```

#### 2020년 1월1일부터 오늘(어제)까지의 영화별 관람객 평점을 가져오자!

```python
movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date) :
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&date={date}'
    html = html.format(date = urllib.parse.quote(today.strftime('%Y%m%d')))
    response = urlopen(html)
    soup = BeautifulSoup(response, 'html.parser')
    
    end = 30
    
    movie_date.extend([today for n in range(0,end)])
    movie_name.extend([soup.find_all('div', 'tit3')[n].a.string for n in range(0,end)])
    
    URL = []
    for n in range(0,end) :
        URL.append(urljoin(html, soup.find_all('div', 'tit3')[n].find('a')['href']))
    
    for n in range(0,end) : 
        html = urlopen(URL[n])
        soup = BeautifulSoup(html, 'html.parser')
        soup = soup.find_all('div', 'star_score')
        
        if soup[0].find(class_ = 'st_off').span.string[-5:-1] == '평점 없' :
            movie_point.append(np.nan)
        elif soup[0].find(class_ = 'st_off').span.string[-5:-1] == '10점 ':
            movie_point.append('10')
        else :
            movie_point.append(soup[0].find(class_ = 'st_off').span.string[-5:-1])
```

```python
movie = pd.DataFrame({'date' : movie_date, 'name' : movie_name, 'point' : movie_point})
movie
```

```python
movie['point'] = movie['point'].astype(float)
```

### 관람객평점 Top10 영화찾기

#### 2020년 1월부터의 매일, 영화의 평점을 평균내어 점수로 환산한다.

#### 10.0은 0.0으로 표기/ 없는 것은 공백/ 개봉예정작은 10으로 표기되어있었음 /평점표자체가 없는경우도10으로 표기되어있음 --> 이들 데이터 다듬기

```python
movie.replace(np.NaN, '관객평없음',regex = True, inplace = True)
movie
```

```python
movie.replace(0.0, 10.0, inplace = True)
movie.replace(10, '관객평없음',inplace = True)
movie
```

```python
movie.replace('관객평없음', 0, inplace = True)
```

```python
movie_unique = pd.pivot_table(movie, index = ['name'], aggfunc = np.mean)
movie_best = movie_unique.sort_values(by = 'point', ascending = False)
movie_best.head(10)
```

##### 찾아보니 img태그에서 8번째 또는 9번째부터  순위가 매겨지기 시작함

```python
movie_rank = []

for today in tqdm_notebook(date) :
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cnt&tg=0&date={date}'
    response = urlopen(html.format(date = urllib.parse.quote(today.strftime('%Y%m%d'))))
    
    soup = BeautifulSoup(response, 'html.parser')
    
    end = 30
    
    for n in range(4,34) :
        if soup.find_all('img')[2*n+1]['alt'] == 'na' or soup.find_all('img')[2*n+1]['alt'] == 'down' or soup.find_all('img')[2*n+1]['alt'] == 'up':
            movie_rank.extend([soup.find_all('img')[2*n]['alt']])
        else :
            movie_rank.extend([soup.find_all('img')[2*n+1]['alt']])
```

```python
movie_rank
```

```python
len(movie_rank)
```

```python
movie['rank'] = movie_rank
movie
```
