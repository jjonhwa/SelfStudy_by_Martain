## 1-1 CCTV 현황과 인구 현황 데이터 구하기

```python
import pandas as pd
import numpy as np
```

```python
CCTV_Seoul = pd.read_excel('../data/~2018_cctv.xlsx', encoding = 'utf-8')
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={'기관명' : '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
CCTV_Seoul['2015년 이전'] = CCTV_Seoul.iloc[:,2:7].sum(axis=1)
CCTV_Seoul.head()
```

```python
CCTV_Seoul.drop(CCTV_Seoul.columns[2:7], axis = 1, inplace = True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_table('../data/2014~2018_population.txt')
pop_Seoul.head()
```

```python
pop_Seoul = pop_Seoul.replace('-','0')
pop_Seoul.to_csv('../data/2014~2018_population.csv')
```

```python
pop_Seoul = pd.read_csv('../data/2014~2018_population.csv', thousands = ',')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns = {'구역별' : '구별', '계' : '인구수'}, inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul.drop(pop_Seoul.columns[0], axis = 1, inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul.drop(pop_Seoul.columns[4:17], axis = 1, inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul['65세이상'] = pop_Seoul.iloc[:,4:12].sum(axis=1)
pop_Seoul.head()
```

```python
pop_Seoul.drop(pop_Seoul.columns[4:12], axis = 1, inplace = True)
pop_Seoul.head()
```

```python
b = pop_Seoul[pop_Seoul['구분'] == '한국인'].iloc[:,3]
b.reset_index(drop = True, inplace = True)
b
```

```python
a = pop_Seoul[pop_Seoul['구분']=='외국인'].iloc[:,3]
a.reset_index(drop=True, inplace = True)
a
```

```python
pop_Seoul.set_index('구분', inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul.drop(['한국인', '외국인'], inplace = True)
pop_Seoul
```

```python
pop_Seoul.reset_index(drop = True, inplace = True)
pop_Seoul
```

```python
pop_Seoul['한국인'] = b
pop_Seoul['외국인'] = a
pop_Seoul.head()
```

```python
pop_Seoul.set_index('기간', inplace = True)
pop_Seoul.drop([2014,2015,2016,2017], inplace = True)
pop_Seoul.head()
```

```python
pop_Seoul.reset_index(drop=True, inplace = True)
pop_Seoul
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['65세이상'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

## 1-8 CCTV 현황 그래프로 분석하기

```python
import matplotlib.pyplot as plt
%matplotlib inline
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin' :
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname = path).get_name()
    rc('font', family = font_name)
else  :
    print('Unknown system... sorry~~~~~')
```

```python
data_result.sort_values(by = '인구수', ascending = False)
```

```python
data_result['소계'].plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
data_result['소계'].sort_values(ascending = True).plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
data_result['CCTV비율'] = data_result['소계'] / data_result['인구수'] * 100
data_result['CCTV비율'].sort_values().plot(kind = 'barh', grid = True, figsize = (10,10))
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s = 50, marker = 'o')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000,700000, 100)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s = 50)
plt.plot(fx, f1(fx),ls = 'dashed', lw = 3, color = 'g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000,700000,100)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s = 50)
plt.plot(fx, f1(fx), ls = 'dashed', lw = 3, color = 'g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by = '오차', ascending = False)
df_sort.head()
```

```python
data_result.dtypes
```

```python
plt.figure(figsize = (12,8))
plt.scatter(data_result['인구수'], data_result['소계'], c = data_result['오차'], s = 50)
plt.plot(fx, f1(fx), ls = 'dashed', lw =  3, color = 'g')

for n in range(10) :
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98, df_sort.index[n], fontsize = 15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar()
plt.grid()
plt.show()
```
