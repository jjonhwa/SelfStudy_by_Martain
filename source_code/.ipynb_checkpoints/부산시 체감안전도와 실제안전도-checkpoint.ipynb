### 부산시 구별 안전도 조사를 통한 안전체감도 조사

```python
safety_sensitivity = pd.read_excel('../data/2018_Busan_Safety_Sensitivity.xlsx', header = 1)
safety_sensitivity.head()
```

```python
safety_sensitivity = safety_sensitivity.iloc[3:19, :]                     #우리가 알고 싶은 구데이터만 가져온다.
safety_sensitivity.reset_index(drop = True, inplace = True)             
safety_sensitivity.drop(columns = '특성별(1)', axis = 1, inplace = True)
safety_sensitivity
```

```python
safety_sensitivity.rename(columns = {'특성별(2)' : '구별', '매우 안전하다' : '매우안전', '비교적 안전하다' : '안전',
                                  '보통이다' : '보통', '비교적 불안하다' : '불안', '매우 불안하다' : '매우불안'}, inplace = True)
safety_sensitivity.head()
```

#### 매우안전부터 순서대로 5,4,3,2,1점을 부여하여 총합 점수로 각 구별로 안전체감도 점수를 매긴다.

```python
safety_sensitivity['안전체감도'] = safety_sensitivity['매우안전'] * 5 + safety_sensitivity['안전']* 4 + \
                            safety_sensitivity['보통'] * 3 + safety_sensitivity['불안'] * 2 + safety_sensitivity['매우불안'] * 1
safety_sensitivity.drop(safety_sensitivity.columns[1:6], axis = 1, inplace = True)
safety_sensitivity.head()
```

```python
safety_sensitivity
```

#### 이후에 경찰서별 범죄 데이터와 구별 안전체감도 데이터를 비교해야하는데
#### 남구경찰서가 남구와 수영구를 관할하므로 안전체감도는 남구와 수영구를 남구로 합치고 점수는 이 둘의 평균으로 함

```python
safety_sensitivity.loc[16] = safety_sensitivity.loc[6] + safety_sensitivity.loc[13]
safety_sensitivity.drop([6,13], axis = 0, inplace = True)
safety_sensitivity.reset_index(drop = True, inplace = True)
safety_sensitivity['구별'][14] = '남구'
safety_sensitivity['안전체감도'][14] /= 2
safety_sensitivity
```

```python
safety_sensitivity.sort_values(by = '구별', ascending = True, inplace = True)
safety_sensitivity.reset_index(drop = True, inplace = True)
safety_sensitivity.head()
```

### 부산시 경찰서별 범죄데이터를 통한 실제안전도 조사

#### 구글맵을 이용해 범죄데이터 다듬기
#### 구글맵에서 경찰서 이름으로 주소를 가져와 이를 구이름으로 변경해주고 싶다

```python
station_name = []

for name in crime_status['경찰서'] :
    if name == '부산진' :                            #부산진구는 이름 앞에 부산이 이미 붙어있으므로 if문 처리해준다.
        station_name.append(name + '경찰서')
    else :
        station_name.append('부산' + name + '경찰서')
        
station_name
```

```python
import googlemaps
gmaps_key = 'AIzaSyAAdqWF_Ykxo8Fcn7FJmrSLfG3LJojryhQ'
gmaps = googlemaps.Client(key = gmaps_key)
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_name :
    tmp = gmaps.geocode(name, language = 'ko')
    station_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc = tmp[0].get('geometry')
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
```

```python
station_address
```

```python
gu_name = []

for name in station_address :
    tmp = name.split()
    gu_name.append(tmp[2])
    
gu_name
```
