```python
si_name = [None] * len(pop)
```

```python
tmp_gu_dict = {'수원' : ['장안구', '권선구', '팔달구', '영통구'],
               '성남' : ['수정구', '중원구', '분당구'],
               '안양' : ['만안구', '동안구'],
               '안산' : ['상록구', '단원구'],
               '고양' : ['덕양구', '일산동구', '일산서구'],
               '용인' : ['처인구', '기흥구', '수지구'],
               '청주' : ['상당구', '서원구', '흥덕구', '청원구'],
               '천안' : ['동남구', '서북구'],
               '전주' : ['완산구', '덕진구'],
               '포항' : ['남구', '북구'],
               '창원' : ['의창구', '성산구', '진해구', '마산합포구', '마산회원구']}
```

#### 2018년 인천광역시의 남구의 명칭이 미추홀구로 바뀌었음

#### 부천시의 3구는 2016년에 폐지되었다

##### 2016년도의 '구'데이터를 얻어와서 명칭을 수정하는 작업을 하였다

```python
for i in pop.index :
    if pop['광역시도'][i][-3:] not in ['특별시', '광역시', '자치시'] :          #이름이 겹치는 고성
        if pop['시군구'][i][:-1] == '고성' and pop['광역시도'][i] == '강원도' :
            si_name[i] = '고성(강원)'
        elif pop['시군구'][i][:-1] == '고성' and pop['광역시도'][i] == '경상남도' :
            si_name[i] = '고성(경남)'  
        else:
            si_name[i] = pop['시군구'][i][:-1]
            
            
        for keys, values in tmp_gu_dict.items():                            #광역시도가 아닌 시의 구이름을 바꿔주기
            if pop['시군구'][i] in values :
                if len(pop['시군구'][i]) == 2 :
                    si_name[i] = keys + ' ' + pop['시군구'][i]
                elif pop['시군구'][i] in ['마산합포구', '마산회원구']:
                    si_name[i] = keys + ' ' + pop['시군구'][i][2:-1]
                else :
                    si_name[i] = keys + ' ' + pop['시군구'][i][:-1]
                  
                
    elif pop['광역시도'][i] == '세종특별자치시' :                               #세종은 자치시라고 되어있지만 그냥 세종으로!
        si_name[i] = '세종'
    
    else :                                                                  # 특별시. 광역시들을 다뤄주자
        if pop['광역시도'][i][-3:] in ['자치시']:                           #제주도의 제주시와 서귀포시도 '시'를 없애주자
            si_name[i] = pop['시군구'][i][:-1]
        elif len(pop['시군구'][i]) == 2:                                   # 특별시,광역시들은 앞의 두글자를 같이 붙여주자
            si_name[i] = pop['광역시도'][i][:2] + ' ' + pop['시군구'][i]
        elif pop['시군구'][i] == '미추홀구' :
            si_name[i] = pop['광역시도'][i][:2] + ' ' + pop['시군구'][i]   #미추홀구는 4글자지만 그냥 사용하도록 하자
        else :
            si_name[i] = pop['광역시도'][i][:2] + ' ' + pop['시군구'][i][:-1]
```

```python
def drawKorea(targetData, blockedMap, cmapname) :
    gamma = 0.75
    
    whitelabelmin = (max(blockedMap[targetData]) - 
                     min(blockedMap[targetData]))*0.25 + \
                     min(blockedMap[targetData])
    
    datalabel = targetData
    
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    
    mapdata = blockedMap.pivot_table(index = 'y', columns = 'x', values = targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize = (9,11))
    plt.pcolor(masked_mapdata, vmin = vmin, vmax = vmax, cmap = cmapname, 
               edgecolor = '#aaaaaa', linewidth = 0.5)

# 지역 이름 표시
    for idx, row in blockedMap.iterrows():
    #광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다.
        #중구, 서구
        if len(row['ID'].split()) == 2 :
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2] == '고성' :
            dispname = '고성'
        else :
            dispname = row['ID']
            
        #서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3 :
            fontsize, linespacing = 10.0, 1.1
        else :
            fontsize, linespacing = 11, 1.
            
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight = 'bold',
                     fontsize = fontsize, ha = 'center', va = 'center', color = annocolor,
                     linespacing = linespacing)
        
        #시도 경계 그리기
    for path in BORDER_LINES :
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)
        
    plt.gca().invert_yaxis()
    
    plt.axis('off')
    
    cb = plt.colorbar(shrink = .1, aspect = 10)
    cb.set_label(datalabel)
    
    plt.tight_layout()
    plt.show()
             
```
